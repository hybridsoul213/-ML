### Задание №4


Ролевое измерение (Role-Playing Dimension) может использоваться по-разному в одной и той же таблице или других таблицах модели данных - это фактически одно и то же измерение, но представленное с разных точек зрения.

* Опиши как ты будешь проектировать ролевое измерение "Дата", которое одновременно используется как "Дата зачисления" и как "Дата завершения" в таблице фактов (**students_events_table_fact в задании №4**) о зачислении студента на курс и успешном завершении студентом курса.

* Напиши SQL-запрос для создания таблицы ролевого измерения "Дата"
```sql
CREATE TABLE dates_dim
```

В данной модели даты зачисления и завершения из фактовой таблицы students_events_table_fact связываются с одним календарным справочником dates_dim. Это ролевое измерение позволяет использовать один справочник для разных целей, обеспечивая непротиворечивость данных.

Справочник спроектирован как готовый инструмент для аналитики, содержащий предрасчитанные поля, которые экономят время и упрощают запросы:

Для группировок: month_name_eng, quarter_num, weekday_name.

Для точных недельных отчётов (WAU): iso_week_num, iso_year_num.

** для фильтрации по периодам:** week_start_date, month_end_date и др. — позволяют легко находить все события за нужный период без сложных условий.

Для анализа аномалий: Признаки is_weekend и is_holiday помогают учитывать влияние выходных и праздников на активность студентов.

Таким образом, dates_dim выступает универсальным и многоразовым компонентом, который поддерживает динамический анализ зачислений, завершений и других метрик без необходимости постоянного пересчёта дат в коде отчётов.

````python
CREATE TABLE dates_dim (
    date_key            INTEGER      PRIMARY KEY,
    full_date           DATE         NOT NULL UNIQUE,

    year_num           INTEGER      NOT NULL,
    quarter_num         INTEGER      NOT NULL,
    month_num          INTEGER      NOT NULL,
    month_name_eng      TEXT         NOT NULL,
    day_of_month        INTEGER      NOT NULL,

    weekday_name        TEXT         NOT NULL,
    weekday_num         INTEGER      NOT NULL,

    iso_week_num        INTEGER      NOT NULL,
    iso_year_num        INTEGER      NOT NULL,
    day_of_week_iso     INTEGER      NOT NULL,

    week_start_date     DATE         NOT NULL,
    week_end_date       DATE         NOT NULL,
    month_start_date    DATE         NOT NULL,
    month_end_date      DATE         NOT NULL,
    quarter_start_date  DATE         NOT NULL,
    quarter_end_date    DATE         NOT NULL,
    year_start_date     DATE         NOT NULL,
    year_end_date       DATE         NOT NULL,

    is_weekend          BOOLEAN      NOT NULL,
    is_holiday          BOOLEAN      DEFAULT FALSE,
    holiday_name        TEXT,

);

INSERT INTO dates_dim (
    date_key,
    full_date,
    year_num,
    quarter_num,
    month_num,
    month_name_eng,
    day_of_month,
    weekday_name,
    weekday_num,
    iso_week_num,
    iso_year_num,
    day_of_week_iso,
    week_start_date,
    week_end_date,
    month_start_date,
    month_end_date,
    quarter_start_date,
    quarter_end_date,
    year_start_date,
    year_end_date,
    is_weekend
)
SELECT
    full_date as d,
    TO_CHAR(d, 'YYYYMMDD')::integer AS date_key,

    EXTRACT(YEAR FROM d)::integer AS year_num,
    EXTRACT(QUARTER FROM d)::integer AS quarter_num,
    EXTRACT(MONTH FROM d)::integer AS month_num,
    TRIM(TO_CHAR(d, 'Month')) AS month_name_eng,
    EXTRACT(DAY FROM d)::integer AS day_of_month,

    TRIM(TO_CHAR(d, 'Day')) AS weekday_name,
    EXTRACT(ISODOW FROM d)::integer AS weekday_num,

    EXTRACT(WEEK FROM d)::integer AS iso_week_num,
    EXTRACT(ISOYEAR FROM d)::integer AS iso_year_num,
    EXTRACT(ISODOW FROM d)::integer AS day_of_week_iso,

    DATE_TRUNC('week', d)::date AS week_start_date,
    (DATE_TRUNC('week', d) + INTERVAL '6 days')::date AS week_end_date,

    DATE_TRUNC('month', d)::date AS month_start_date,
    (DATE_TRUNC('month', d) + INTERVAL '1 month - 1 day')::date AS month_end_date,

    DATE_TRUNC('quarter', d)::date AS quarter_start_date,
    (DATE_TRUNC('quarter', d) + INTERVAL '3 months - 1 day')::date AS quarter_end_date,

    DATE_TRUNC('year', d)::date AS year_start_date,
    (DATE_TRUNC('year', d) + INTERVAL '1 year - 1 day')::date AS year_end_date,

    EXTRACT(ISODOW FROM d)::integer IN (6, 7) AS is_weekend

FROM GENERATE_SERIES(
    '2018-01-01'::date,
    '2030-12-31'::date,
    INTERVAL '1 day'
) AS gs(d)
ORDER BY d;
````

````python
# СЛОЙ ИНФРАСТРУКТУРЫ

class DatabaseRepository:

    def __init__(self, db_type='clickhouse'):
        self.db_type = db_type
        self.client = None
        self._connect()

    def _connect(self):
        try:
            if self.db_type == 'postgresql':
                import psycopg2
                self.client = psycopg2.connect(
                    "dbname='db' user='user' password='password' host='localhost'"
                )
            elif self.db_type == 'mysql':
                import pymysql
                self.client = pymysql.connect(
                    host='localhost',
                    user='user',
                    password='password',
                    database='db',
                    charset='utf8mb4'
                )
            elif self.db_type == 'clickhouse':
                import clickhouse_connect
                self.client = clickhouse_connect.get_client(
                    host='localhost',
                    username='user',
                    password='password'
                )
            else:
                raise ValueError(f"Unsupported database type: {self.db_type}")

        except ImportError as e:
            raise Exception(f"Database driver not installed: {e}")

    def execute_query(self, query: str, params=None):
        if self.db_type == 'clickhouse':
            # Для ClickHouse оптимизируем запрос
            optimized_query = query.replace("WHERE", "PREWHERE")
            result = self.client.query(optimized_query, params).result_rows
        else:
            # Для PostgreSQL/MySQL
            with self.client.cursor() as cursor:
                cursor.execute(query, params or {})
                result = cursor.fetchall()

        return result

    def close(self):
        if self.client:
            self.client.close()


# СЛОЙ ПРИЛОЖЕНИЯ

class ARPUService:

    def __init__(self, repository: DatabaseRepository):
        self.repository = repository

    def calculate_daily_metrics(self, target_date: str = None):
        import datetime

        if target_date is None:
            yesterday = datetime.datetime.now() - datetime.timedelta(days=1)
            target_date = yesterday.strftime('%Y-%m-%d')

        query = """
            SELECT
                to_char(eventtime, 'YYYY-MM-DD') as "date",
                SUM(priceRUB) as "revenue",
                COUNT(user_id) as "enrollments_count",
                COUNT(DISTINCT user_id) as "unique_students",
                AVG(priceRUB) as "average_check",
                SUM(priceRUB) / NULLIF(COUNT(DISTINCT user_id), 0) as "arpu"
            FROM myproduct.payments
            WHERE eventtime >= %(start_date)s
                AND eventtime < %(end_date)s
            GROUP BY to_char(eventtime, 'YYYY-MM-DD')
        """

        params = {
            'start_date': target_date,
            'end_date': (datetime.datetime.strptime(target_date, '%Y-%m-%d')
                        + datetime.timedelta(days=1)).strftime('%Y-%m-%d')
        }

        if isinstance(self.repository, DatabaseRepository) and self.repository.db_type == 'clickhouse':
            query = query.replace("WHERE", "PREWHERE")

        try:
            result = self.repository.execute_query(query, params)
            return self._format_metrics(result, target_date)
        except Exception as e:
            raise Exception(f"Error calculating daily metrics: {e}")

    def _format_metrics(self, raw_data, target_date):
        if not raw_data:
            return {
                'date': target_date,
                'revenue': 0,
                'enrollments_count': 0,
                'unique_students': 0,
                'average_check': 0,
                'arpu': 0,
                'status': 'no_data'
            }

        if isinstance(raw_data[0], (list, tuple)):
            return {
                'date': raw_data[0][0],
                'revenue': float(raw_data[0][1] or 0),
                'enrollments_count': int(raw_data[0][2] or 0),
                'unique_students': int(raw_data[0][3] or 0),
                'average_check': float(raw_data[0][4] or 0),
                'arpu': float(raw_data[0][5] or 0),
                'status': 'success'
            }
        else:
            return {
                'date': raw_data[0]['date'],
                'revenue': float(raw_data[0]['revenue'] or 0),
                'enrollments_count': int(raw_data[0]['enrollments_count'] or 0),
                'unique_students': int(raw_data[0]['unique_students'] or 0),
                'average_check': float(raw_data[0]['average_check'] or 0),
                'arpu': float(raw_data[0]['arpu'] or 0),
                'status': 'success'
            }


class MetricsReporter:

    def __init__(self, arpu_service: ARPUService):
        self.arpu_service = arpu_service

    def generate_daily_report(self):
        metrics = self.arpu_service.calculate_daily_metrics()

        print("=" * 50)
        print("ЕЖЕДНЕВНЫЙ ОТЧЕТ ПО ARPU")
        print("=" * 50)
        print(f"Дата: {metrics['date']}")
        print(f"Выручка: {metrics['revenue']:,.2f} руб.")
        print(f"Количество зачислений: {metrics['enrollments_count']}")
        print(f"Уникальных студентов: {metrics['unique_students']}")
        print(f"Средний чек: {metrics['average_check']:,.2f} руб.")
        print(f"ARPU: {metrics['arpu']:,.2f} руб.")
        print("=" * 50)

        return metrics


def main():
    try:
        repository = DatabaseRepository(db_type='clickhouse')

        arpu_service = ARPUService(repository)
        reporter = MetricsReporter(arpu_service)
        report = reporter.generate_daily_report()

        return report

    except Exception as e:
        print(f"Ошибка при выполнении: {e}")
        return None
    finally:
        if 'repository' in locals():
            repository.close()


if __name__ == "__main__":
    main()
````
